```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#calculating total unique number of players and creating a dataframe to represent the value
total_players = purchase_data['SN'].nunique()
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculations for Purchasing Analysis and providing the totals in a table
unique_items = purchase_data['Item Name'].nunique()
average_price = round(purchase_data['Price'].mean(),2)
number_purchases = purchase_data['Purchase ID'].count()
total_revenue = round((average_price * number_purchases), 2)

summary_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                          "Average Price":[average_price],
                          "Number of Purchases":[number_purchases],
                          "Total Revenue":"$"+ str(total_revenue)})

#formatting/styling the dataframe
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Calculations for Gender Demographics.
#Creating a list of columns to pass to the loc function to create gender based dataframes
columns = [
    "Purchase ID",
    "SN",
    "Age",
    "Gender",
    "Item ID",
    "Item Name",
    "Price",
]

#creating a dataframe per gender
male_df = purchase_data.loc[purchase_data["Gender"]=="Male", columns]

female_df = purchase_data.loc[purchase_data["Gender"]=="Female", columns]

other_df = purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed", columns]
```

```python
#calculating the percentage and count of gender players per dataframe
unique_maleplayers = male_df['SN'].nunique()
percentage_male = round((unique_maleplayers/total_players)*100, 2)

unique_femaleplayers = female_df['SN'].nunique()
percentage_female = round((unique_femaleplayers/total_players)*100, 2)

unique_otherplayers = other_df['SN'].nunique()
percentage_other = round((unique_otherplayers/total_players)*100, 2)

#creating lists to create dataframe
gender = ["Male","Female","Other / Non-Disclosed"]
total_count = [unique_maleplayers, unique_femaleplayers, unique_otherplayers]
per_players = [percentage_male, percentage_female, percentage_other]
```

```python
gender_demo_df = pd.DataFrame({"Gender": gender,
                               "Total Count": total_count,
                               "Percentage of Players": per_players
                               })
gender_demo_df

#Setting index by Gender
gender_demo_df.set_index('Gender')
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# defining a function to calculate the values for each dataframe (gender based)
def DFcalculation(df):
    pur_count = df["Purchase ID"].count()
    avg_price = round(df["Price"].mean(),2)
    total_Pur = round(df["Price"].sum(),2)
    unique_person = df["SN"].nunique()
    avg_pur_total = round(total_Pur/unique_person,2)
    return (pur_count, avg_price, total_Pur, avg_pur_total)

#calling the above function to create a summary series per gender
female_summary = DFcalculation(female_df)
male_summary = DFcalculation(male_df)
other_summary = DFcalculation(other_df)
gender = ["Female", "Male","Other / Non-Disclosed"]

#creating lists from each elment of the above data series to pass list for creating the dataframe
purcount_list = [female_summary[0], male_summary[0], other_summary[0]]
avgpur_list = [female_summary[1], male_summary[1], other_summary[1]]
totpur_list = [female_summary[2], male_summary[2], other_summary[2]]
avg_pur_tot_list = [female_summary[3], male_summary[3], other_summary[3]]


pur_analysis_df = pd.DataFrame({"Gender": gender,
                                "Purchase Count": purcount_list,
                               "Average Purchase Price": avgpur_list,
                               "Total Purchase Value": totpur_list,
                                "Avg Total Purchase per Person": avg_pur_tot_list

                               })

#formatting/styling the dataframe
pur_analysis_df["Average Purchase Price"] = pur_analysis_df["Average Purchase Price"].map("${:.2f}".format)
pur_analysis_df["Total Purchase Value"] = pur_analysis_df["Total Purchase Value"].map("${:.2f}".format)
pur_analysis_df["Avg Total Purchase per Person"] = pur_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#Setting index by Gender
pur_analysis_df.set_index('Gender')
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#creating a smaller dataframe of unique Screen names and Age columns to calculate the age demographics values
reduced_purchase_data = purchase_data.loc[:, ['SN', 'Age']]
unique_reduced_purchase_data = reduced_purchase_data.drop_duplicates()
```

```python
#creating bins for years
bins = [0, 9, 14, 19, 24, 29, 34, 39, 44]

# Create the names/labels for the bins
group_age = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

unique_reduced_purchase_data["Age Range"] = pd.cut(unique_reduced_purchase_data["Age"], bins, labels = group_age)

#calculating the count of players
df1=unique_reduced_purchase_data["Age Range"].value_counts()

#calculating the percentage of players
df2=round(unique_reduced_purchase_data["Age Range"].value_counts()*100/unique_reduced_purchase_data["Age Range"].count(),2)

```

```python
age_demo_df = pd.DataFrame({"Total Count": df1,
                           "Percentage of players": df2})
age_demo_df
age_demo_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#creating bins for years
bins = [0, 9, 14, 19, 24, 29, 34, 39, 44]

# Create the names/labels for the bins
group_age = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_age)
grouped_purchase = purchase_data.groupby("Age Range")
```

```python
pur_count_age = grouped_purchase["Purchase ID"].count()
avg_price_age = round(grouped_purchase["Price"].mean(),2)
total_Pur_age = round(grouped_purchase["Price"].sum(),2)
unique_person_age = grouped_purchase["SN"].nunique()
avg_pur_total_age = round(total_Pur_age/unique_person_age,2)
    

total_pur_analysis = pd.DataFrame({"Purchase Count": pur_count_age,
                                    "Average Purchase Price": avg_price_age,
                                    "Total Purchase Value": total_Pur_age,
                                    "Avg Total Purchase per Person": avg_pur_total_age
                                    })
total_pur_analysis["Average Purchase Price"] = total_pur_analysis["Average Purchase Price"].map("${:.2f}".format)
total_pur_analysis["Total Purchase Value"] = total_pur_analysis["Total Purchase Value"].map("${:.2f}".format)
total_pur_analysis["Avg Total Purchase per Person"] = total_pur_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

total_pur_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_purchase_TS = purchase_data.groupby("SN")
grouped_purchase_TS
purchase_count_TS = grouped_purchase_TS["Purchase ID"].count()

avg_pur_price_TS = round(grouped_purchase_TS["Price"].mean(),2)

tot_pur_value_TS = round(grouped_purchase_TS["Price"].sum(),2)

spender_df = pd.DataFrame({"Purchase Count":purchase_count_TS,
                              "Average Purchase Price": avg_pur_price_TS,
                              "Total Purchase Value":tot_pur_value_TS})

#top_spender_df.style.format({"Average Purchase Price": "${:,.2f}",
                           #"Total Purchase Value": "${:,.2f}"})
    
top_spender_df = spender_df.nlargest(5, 'Total Purchase Value')

top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)

top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_purchase_PI = purchase_data.groupby(["Item ID","Item Name"])
grouped_purchase_PI

purchase_count_PI = grouped_purchase_PI["Item Name"].count()

tot_pur_value_PI = round(grouped_purchase_PI["Price"].sum(),2)

item_price_PI = tot_pur_value_PI/purchase_count_PI

popular_items_df = pd.DataFrame({"Purchase Count":purchase_count_PI,
                              "Item Price": item_price_PI,
                              "Total Purchase Value":tot_pur_value_PI})

top_items_df = popular_items_df.nlargest(5, 'Purchase Count')

top_items_df["Item Price"] = top_items_df["Item Price"].map("${:.2f}".format)
top_items_df["Total Purchase Value"] = top_items_df["Total Purchase Value"].map("${:.2f}".format)

top_items_df
```

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting

```python
top_profit_df = popular_items_df.nlargest(5, 'Total Purchase Value')

top_profit_df["Item Price"] = top_profit_df["Item Price"].map("${:.2f}".format)
top_profit_df["Total Purchase Value"] = top_profit_df["Total Purchase Value"].map("${:.2f}".format)

top_profit_df
```
